```python
import numpy as np

years = 5

# Conventional
conv_initial = 50000
conv_annuals = np.array([80000] * years)
conv_cum = np.cumsum(np.concatenate([[-conv_initial],
conv_annuals]))

# Organic
org_initial = 70000
org_annuals = np.array([95000] * years)
org_annuals[2] -= 5000  # Year 3 certification
org_cum = np.cumsum(np.concatenate([[-org_initial],
org_annuals]))

# Outputs
print("Conventional cumulative:", conv_cum)
print("Organic cumulative:", org_cum)
print("Conv. 5-yr profit:", conv_cum[-1])
print("Org. 5-yr profit:", org_cum[-1])
```
